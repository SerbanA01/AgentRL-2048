```python
def is_valid_move(env, action):

    original_board = env.get_board().copy()

    try:
        # Perform the action
        env.step(action)
        # Check if the board has changed
        board_changed = not np.array_equal(env.get_board(), original_board)
        # Revert the board to the original state
        env.set_board(original_board)
        return board_changed
    except IllegalMove:
        return False
```

```python
  # Create the 2048 environment
  env = gym.make('2048-v0')

  def random_agent_with_stats(env, episodes=10000):
      def has_valid_moves(env):
          """Check if there are any valid moves available."""
          for action in range(4):  # Actions: 0 (up), 1 (right), 2 (down), 3 (left)
              if is_valid_move(env, action):
                  return True
          return False

      # Statistics
      scores = []
      highest_tiles = []
      episode_lengths = []

      for episode in range(episodes):
          obs = env.reset()  # Reset the environment
          total_score = 0
          moves = 0

          while True:

              env.render(mode='human')
              # Check if there are any valid moves left
              if not has_valid_moves(env):
                  break  # End the game if no valid moves are left

              # Choose a random valid move
              while True:
                  action = env.action_space.sample()
                  if is_valid_move(env, action):  # Ensure the action is valid
                      obs, reward, done, info = env.step(action)
                      total_score += reward
                      moves += 1
                      break  # Break once a valid move is made

          # Collect statistics
          scores.append(total_score)
          highest_tiles.append(env.highest())
          episode_lengths.append(moves)

      # Analyze results
      average_score = np.mean(scores)
      max_score = np.max(scores)
      average_moves = np.mean(episode_lengths)

      # Highest Tile Distribution
      unique_tiles, counts = np.unique(highest_tiles, return_counts=True)

      # Display statistics
      print(f"--- Random Agent Statistics over {episodes} Episodes ---")
      print(f"Average Score: {average_score}")
      print(f"Max Score: {max_score}")
      print(f"Average Moves per Episode: {average_moves}")
      print("Highest Tile Distribution:")
      for tile, count in zip(unique_tiles, counts):
          print(f"Tile {tile}: {count} times ({(count / episodes) * 100:.2f}%)")

      # Visualizations
      # 1. Score Distribution
      plt.figure(figsize=(10, 6))
      plt.hist(scores, bins=20, edgecolor='black')
      plt.title('Score Distribution Across Episodes')
      plt.xlabel('Score')
      plt.ylabel('Frequency')
      plt.grid(axis='y', alpha=0.75)
      plt.show()

      # 2. Highest Tile Distribution
      plt.figure(figsize=(10, 6))
      plt.bar(unique_tiles, counts, color='skyblue', edgecolor='black')
      plt.title('Highest Tile Distribution')
      plt.xlabel('Highest Tile Achieved')
      plt.ylabel('Frequency')
      plt.xticks(unique_tiles)
      plt.grid(axis='y', alpha=0.75)
      plt.show()

      # 3. Moves per Episode Distribution
      plt.figure(figsize=(10, 6))
      plt.hist(episode_lengths, bins=20, edgecolor='black')
      plt.title('Moves per Episode Distribution')
      plt.xlabel('Number of Moves')
      plt.ylabel('Frequency')
      plt.grid(axis='y', alpha=0.75)
      plt.show()

      # 4. Score vs. Number of Games
      plt.figure(figsize=(10, 6))
      plt.plot(range(1, episodes + 1), scores, marker='o', linestyle='-', markersize=4, label='Score')
      plt.title('Score vs. Number of Games')
      plt.xlabel('Game Number')
      plt.ylabel('Score')
      plt.grid(alpha=0.75)
      plt.legend()
      plt.show()
  # Run the random agent
  random_agent_with_stats(env)
```
